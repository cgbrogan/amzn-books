```python
import pandas as pd
import numpy as np
import time

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# Title: Ride of a Lifetime
# Author: Robert Iger
# Rating: 4.8 / 5
# Reviews: 1570

url = 'https://www.amazon.com/Ride-Lifetime-Lessons-Learned-Company-ebook/dp/B07PF6XTD8'
html = urlopen(url)
soup = BeautifulSoup(html, 'html')
```

```python
soup
```

```python
title = soup.find(id='ebooksProductTitle').string.strip()
rating = soup.find("span", class_="a-icon-alt").string.strip()
author = soup.find("a", class_="a-link-normal contributorNameID").string.strip()
number_of_reviews = soup.find(id='acrCustomerReviewText').string.strip()

print(title, rating, author, number_of_reviews)
```
